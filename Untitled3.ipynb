```python
#preprocessing 
# 1.0 resizing 
# 2.0 flattern

import os
import cv2 
import skimage
from sklearn import preprocessing as p
import matplotlib.pyplot as plt 
import numpy as np
from skimage.io import imread
from skimage.transform import resize

target = []
images = []
flat_data = []

DATADIR = '/content/images'
CATEGORIES = ['overripe_oranges','ripe_oranges','unripe_oranges']

for category in CATEGORIES :
  class_num = CATEGORIES.index(category) # label encoding 
  path = os.path.join(DATADIR, category) #creating path 
  
  for img in os.listdir(path) :
    new = os.path.join(path,img)
    if new.endswith('.jpg'):

      img_array = imread(new,plugin='matplotlib')
      img_array = np.array(img_array)
      
      # print(img_array.shape)
      # plt.imshow(img_array)
      # break
      
      img_resized = resize(img_array,(150,150,3))
      flat = img_resized.flatten()
      flat_data.append(flat) 
      images.append(img_resized)
      target.append(class_num)

target = np.array(target)
images = np.array(images)
flat_data = np.array(flat_data)


```

```python
unique,count = np.unique(target ,return_counts=True)
plt.bar(CATEGORIES,count)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test ,y_train,y_test = train_test_split(flat_data , target , test_size =0.3 ,
                                                  random_state=109)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

param_grid = [
              { 'C':[1,10,100,1000] ,'kernel' :['linear']},
              { 'C':[1,10,100,1000] ,'gamma':[0.001,0.0001] ,'kernel' :['rbf']},
                  
]
svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
#save the model 
import pickle
pickle.dump(clf,open('img_model.p','wb'))
model = pickle.load(open('img_model.p','rb'))
```
